```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
from scipy.stats import shapiro
from scipy import stats
```

```python
df = pd.read_excel('21.10 CASE_LOSSES.xlsx')
df
```

```python
df2 = pd.read_excel('21.10 CASE_CONTRACTS.xlsx')
df2
```

```python
df3 = pd.read_excel('21.10 CASE_CLIENTS.xlsx')
df3.head()
```

```python
df4 = pd.read_excel('21.10 experiment_group.xlsx')
df4
```

```python
result_df = df4.join(other=df2.set_index('client_id'), on='client_id')
result_df
```

```python
result_df1 = result_df.join(other=df.set_index('client_id'), on='client_id')
result_df1
```

```python
del result_df1['Unnamed: 4']
result_df1
```

```python
del result_df1['Unnamed: 5']
result_df1
```

```python
# Курс обмена
exchange_rate = 80

# Применение лямбда-функции для конвертации данных по страховуй сумме
result_df1['insurance_amount'] = result_df1.apply(lambda row: row['insurance_amount'] / exchange_rate if row['currency_name'] == 'Российский рубль' else row['insurance_amount'], axis=1)

result_df1
```

```python
# Курс обмена
exchange_rate = 80

# Применение лямбда-функции для конвертации данных по страховуй сумме
result_df1['price'] = result_df1.apply(lambda row: row['price'] / exchange_rate if row['currency_name'] == 'Российский рубль' else row['price'], axis=1)

result_df1
```

```python
# Курс обмена
exchange_rate = 80

# Применение лямбда-функции для конвертации данных по страховуй сумме
result_df1['loss_payout_amt'] = result_df1.apply(lambda row: row['loss_payout_amt'] / exchange_rate if row['currency_name'] == 'Российский рубль' else row['loss_payout_amt'], axis=1)

result_df1
```

```python
result_df1['group'].unique()
```

```python
df6 = result_df1[result_df1['group'] == 'control']
df6
```

```python
df6_1 = df6[df6['price'] > 0]
df6_1 # датафрейм для контрольной группы, которые заключили договор страхования
```

```python
df7 = result_df1[result_df1['group'] == 'test']
df7
```

```python
df7_1 = df7[df7['price'] > 0]
df7_1  # датафрейм для тестовой группы, которые заключили договор страхования
```

```python
# Проводим T-тест
stat, p = ttest_ind(df6_1['price'], df7_1['price'])

print('Статистика T =', stat)
print('p-значение =', p)

if p < 0.05:
    print('Отвергаем нулевую гипотезу: средние значения выборок статистически значимо различаются.')
else:
    print('Не отвергаем нулевую гипотезу: нет статистически значимых различий между средними значениями выборок.')
```

```python
# Проводим тест Шапиро-Уилка
stat, p_value = shapiro(df6_1['price'])

print('Статистика теста:', stat)
print('p-значение:', p_value)

alpha = 0.05
if p_value > alpha:
    print('Данные можно считать нормально распределенными')
else:
    print('Данные не являются нормально распределенными')
```

```python
# Проводим тест Шапиро-Уилка
stat, p_value = shapiro(df7_1['price'])

print('Статистика теста:', stat)
print('p-значение:', p_value)

alpha = 0.05
if p_value > alpha:
    print('Данные можно считать нормально распределенными')
else:
    print('Данные не являются нормально распределенными')
```

```python
# Проводим тест Манна — Уитни для не нормально распределенных выборок при альтернативной гипотезе , что выботки отличаются
stat5, p5 = stats.mannwhitneyu(df6_1['price'], df7_1['price'])

print('Статистика T =', stat5)
print('p-значение =', p5)

if p5 < 0.05:
    print('Отвергаем нулевую гипотезу: средние значения выборок статистически значимо различаются.')
else:
    print('Не отвергаем нулевую гипотезу: нет статистически значимых различий между средними значениями выборок.')
```
